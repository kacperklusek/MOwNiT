```python
import numpy as np
import matplotlib.pyplot as plt
import math

import pandas as pd

from scipy.special import roots_chebyt

pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

## Funkcja interpolujca zgodnie z poniszym wzorem (Lagrange)


```python
def Lk(X, xk, k, x_points):
    d = m = 1
    for i in range(len(x_points)):
        if i!= k:
            d *= (X - x_points[i])
    for i in range(len(x_points)):
        if i != k:
            m *= (xk - x_points[i])
    return d/m

def lagrange_interpolation(X, x, y):
    result = 0
    for k in range(len(x)):
        result += y[k] * Lk(X, x[k], k, x)
    return result
```

## Funkcja interpolujca zgodnie z motod Newtona z ilorazami rnicowymi

```python
def newton_interpolation(xi, x, y):
    n = len(x)
    div_der = [[None for x in range(n)] for x in range(n)]
    yint = [None for x in range(n)]
    
    for i in range(n):
        div_der[i][0] = y[i]
    for j in range(1, n):
        for i in range(n-j):
            div_der[i][j] = (div_der[i+1][j-1] - div_der[i][j-1])/(x[i+j]-x[i])
    
    xterm = 1
    yint[0] = div_der[0][0]
    for o in range(1, n):
        xterm *= (xi - x[o-1])
        yint[o] = yint[o-1] + div_der[0][o]*xterm
        
    return yint[-1]
```

### funkcja rysujca wykresy

```python
def plot(space, *functions, points=False, title=False):
    if points != False:
        plt.scatter(points[0], points[1], label="nodes")
    
    for foo, lbl in functions:   
        plt.plot(x_space, foo(space), label=lbl)
    
    if title:    
        plt.title(title)
        
    plt.rcParams['figure.figsize'] = [10, 7]
    plt.legend()
    plt.grid()
    plt.show()
```

# Funkcja zadana - zadanie indywidualne


```python
k = 1
m = 2
def f(x):
    return -k * x * math.sin(m*(x-1))

f = np.vectorize(f)
```

### przedzia [-3pi+1, 2pi+1]

```python
a = -3 * math.pi + 1
b = 2 * math.pi + 1
```

### punkty do rysowania wykresu

```python
x_space = np.linspace(a, b, 1000)
```

```python
def chebyshev(a, b, n):
    x, weights = roots_chebyt(n)
    scale = abs(a-b) / 2
    shift = (b+a) / 2
    x *= scale
    x += shift
    return x
```

```python
def equidistant(a, b, n):
    x = np.linspace(a, b, n)
    return x
```

## f(x)

```python
plot(x_space, (f, "f(x)"))
```

## Metryki bdu / dokadnoci funkcji

```python
def diff_max(X1, X2):
    return max([abs(X1[i] - X2[i]) for i in range(len(X1))])
```

```python
def diff_SSE(X1, X2):
    return sum([(X1[i] - X2[i]) ** 2 for i in range(len(X1))])
```

## Mierzenie dokadnoci funkcji interpolowanej dla n wzw

### lagrange (chebyshev)

```python
df = pd.DataFrame(columns=["n", "SSE", "MAX"])

for n in [3,4,5,7,10,15,20,25,55,100]:
    x_ch = chebyshev(a, b, n)
    cheb_lagrange = lagrange_interpolation(x_space, x_ch, f(x_ch))
    
    dsse = diff_SSE(cheb_lagrange, f(x_space))
    dmax = diff_max(cheb_lagrange, f(x_space))
    
    df.loc[len(df.index)] = [n, dsse, dmax]
    
    plot(x_space, 
         (lambda x: lagrange_interpolation(x_space, x_ch, f(x_ch)), "interpolated poly"),
            (f, 'f(x)'), 
            points = [x_ch, f(x_ch)],
            title=f"lagrange - chebyshew  (n={n})\n\
            DIFF_MAX: {dmax:.5f} \
            DIFF_SSE: {dsse:.5f}")
```

```python
df
```

    dla funkcji lagrange z wzami okrelonymi zerami wielomianu czebyszewa 
    wiksze n zawsze daje wiksz dokadno

### lagrange (rwnoodlege)

```python
df = pd.DataFrame(columns=["n", "SSE", "MAX"])

for n in [3,5,7,10,20,25,55,75]:
    x_eq = equidistant(a, b, n)
    equi_lagrange = lagrange_interpolation(x_space, x_eq, f(x_eq))
    
    dmax = diff_max(equi_lagrange, f(x_space))
    dsse = diff_SSE(equi_lagrange, f(x_space))
    
    df.loc[len(df.index)] = [n, dsse, dmax]
    
    plot(x_space, 
         (lambda x: lagrange_interpolation(x_space, x_eq, f(x_eq)), "interpolated poly"),
            (f, 'f(x)'), 
            points = [x_eq, f(x_eq)],
            title=f"lagrange - equidistant  (n={n})\n\
            DIFF_MAX: {dmax:.5f} \
            DIFF_SSE: {dsse:.5f}")   
```

### Newton (chebychev)

```python
df = pd.DataFrame(columns=["n", "SSE", "MAX"])

for n in [3,4,5,7,10,15,20,25,75]:
    x_ch = chebyshev(a, b, n)
    cheb_newton = newton_interpolation(x_space, x_ch, f(x_ch))
    
    dsse = diff_SSE(cheb_newton, f(x_space))
    dmax = diff_max(cheb_newton, f(x_space))
    
    df.loc[len(df.index)] = [n, dsse, dmax]
    
    plot(x_space, 
         (lambda x: newton_interpolation(x_space, x_ch, f(x_ch)), "interpolated poly"),
            (f, 'f(x)'), 
            points = [x_ch, f(x_ch)],
            title=f"newton - chebyshew  (n={n})\n\
            DIFF_MAX: {dmax:.5f} \
            DIFF_SSE: {dsse:.5f}")
```
