### W celu uzyskania wysokiej precyzji wykonywanych operacji na liczbach zmiennoprzecinkowych, używam biblioteki  ```decimal```

````python
import decimal
from decimal import Decimal

import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
````

#### Pierwsze k liczb ciągu

````python
PRECISION = 64
k = 170

precise = []

with decimal.localcontext() as ctx:
    x_0 = Decimal('1') / Decimal('10')
    
    ctx.prec = PRECISION
    ctx.rounding = decimal.ROUND_HALF_EVEN
    
    x_k = x_0
    precise.append(float(x_0))
    
    for i in range(1, k+1):
        x_k = x_k + 3 * x_k * (1 - x_k)
        precise.append(float(x_k))
````

````python
data = pd.DataFrame(columns=['prec'])
data['prec'] = precise
````

````python
sns.set_theme()
plt.figure(figsize=(11,7))
sns.lineplot(data=data['prec'], linewidth=2)
plt.show()
````
